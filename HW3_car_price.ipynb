# Q1

```python
for c in ['make', 'model', 'engine_fuel_type', 'transmission_type', 'driven_wheels', 'market_category', 'vehicle_size', 'vehicle_style']:
    car[c] = car[c].replace(to_replace=np.nan, value="unk")
for c in [ 'year','engine_hp','engine_cylinders', 'number_of_doors', 'highway_mpg', 'city_mpg', 'popularity']:
    car[c] = car[c].replace(to_replace=np.nan, value=0)
```

```python
mean_price = car['price'].mean()

car['above_average'] = (car['price'] > mean_price).astype(int)
```

```python
print(car)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(car, test_size=0.2, random_state=1)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
y_train = df_train.reset_index(drop=True)
y_val = df_val.reset_index(drop=True)
y_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.price.values
y_val = df_val.price.values
y_test = df_test.price.values
```

# Mutual info

```python
from sklearn.metrics import mutual_info_score
```

```python
score = mutual_info_score(df_full_train.above_average, df_full_train.transmission_type)
round(score, 2)
```

```python
score = mutual_info_score(df_full_train.above_average, df_full_train.make)
round(score, 2)
```

```python
score = mutual_info_score(df_full_train.above_average, df_full_train.vehicle_style)
round(score, 2)
```

```python
score = mutual_info_score(df_full_train.above_average, df_full_train.model)
round(score, 2)
```

The variables with the lowest mutual information score is: transmission_type
